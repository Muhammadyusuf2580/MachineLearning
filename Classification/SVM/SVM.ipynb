```python
# Let's create a sample dataset with 50 sentences for each of 3 languages: English, Uzbek, Russian.

# English sample sentences
english_sentences = [
    "How are you?", "I am fine.", "What is your name?", "Where do you live?", "I love programming.",
    "This is a beautiful day.", "The weather is nice today.", "Can you help me?", "I am learning English.",
    "She is my friend.", "He goes to school.", "Do you like pizza?", "We are going to the market.",
    "I need some water.", "Turn on the light.", "This book is interesting.", "Please close the door.",
    "What time is it?", "I will call you later.", "It is raining outside.", "The cat is on the roof.",
    "I like watching movies.", "He is playing football.", "They are reading books.", "Do you speak English?",
    "We are having dinner.", "Open the window.", "It is cold today.", "I am tired.", "Let's go shopping.",
    "I have a dog.", "She has a bicycle.", "The sun is shining.", "Do you want some tea?",
    "This is my house.", "I lost my keys.", "He is a good teacher.", "Where is the bathroom?",
    "Take care!", "Nice to meet you.", "I am hungry.", "The bus is late.", "Turn left at the corner.",
    "You are very kind.", "We are happy.", "I am going home.", "The cake tastes good.",
    "He is my brother.", "She is very smart.", "This is my favorite song.", "I enjoy reading novels."
]

# Uzbek sample sentences
uzbek_sentences = [
    "Qanday yashayapsiz?", "Men yaxshiman.", "Ismingiz nima?", "Qayerda yashaysiz?", "Men dasturlashni yaxshi ko‘raman.",
    "Bugun chiroyli kun.", "Bugun ob-havo yaxshi.", "Menga yordam bera olasizmi?", "Men ingliz tilini o‘rganayapman.",
    "U mening do‘stim.", "U maktabga boradi.", "Pitsani yoqtirasizmi?", "Biz bozorga ketyapmiz.",
    "Menga suv kerak.", "Chiroqni yoqing.", "Bu kitob qiziqarli.", "Iltimos, eshikni yoping.",
    "Soat nechchi bo‘ldi?", "Keyinroq qo‘ng‘iroq qilaman.", "Tashqarida yomg‘ir yog‘ayapti.",
    "Mushuk tomda yuribdi.", "Men kino ko‘rishni yoqtiraman.", "U futbol o‘ynamoqda.",
    "Ular kitob o‘qiyapti.", "Ingliz tilida gaplashasizmi?", "Biz kechki ovqat qilyapmiz.",
    "Derazani oching.", "Bugun sovuq.", "Men charchadim.", "Bozorga boraylik.",
    "Menda it bor.", "Unda velosiped bor.", "Quyosh porlamoqda.", "Choy ichasizmi?",
    "Bu mening uyim.", "Kalitlarimni yo‘qotdim.", "U yaxshi o‘qituvchi.", "Hammom qayerda?",
    "O‘zingizni ehtiyot qiling!", "Siz bilan tanishganimdan xursandman.", "Men ochman.",
    "Avtobus kechikdi.", "Burchakda chapga buriling.", "Siz juda mehribonsiz.",
    "Biz xursandmiz.", "Uyga ketayapman.", "Tort mazali ekan.", "U mening akam.",
    "U juda aqlli.", "Bu mening sevimli qo‘shig‘im.", "Roman o‘qishni yoqtiraman."
]

# Russian sample sentences
russian_sentences = [
    "Как дела?", "Я в порядке.", "Как тебя зовут?", "Где ты живешь?", "Я люблю программировать.",
    "Сегодня прекрасный день.", "Погода сегодня хорошая.", "Ты можешь мне помочь?", "Я изучаю английский язык.",
    "Она моя подруга.", "Он ходит в школу.", "Ты любишь пиццу?", "Мы идем на рынок.",
    "Мне нужна вода.", "Включи свет.", "Эта книга интересная.", "Пожалуйста, закрой дверь.",
    "Который час?", "Я перезвоню позже.", "На улице идет дождь.", "Кошка на крыше.",
    "Мне нравится смотреть фильмы.", "Он играет в футбол.", "Они читают книги.",
    "Ты говоришь по-английски?", "Мы ужинаем.", "Открой окно.", "Сегодня холодно.",
    "Я устал.", "Пойдем по магазинам.", "У меня есть собака.", "У нее есть велосипед.",
    "Светит солнце.", "Хочешь чаю?", "Это мой дом.", "Я потерял ключи.",
    "Он хороший учитель.", "Где ванная комната?", "Береги себя!", "Приятно познакомиться.",
    "Я голоден.", "Автобус опаздывает.", "Поверни налево на углу.", "Ты очень добрый.",
    "Мы счастливы.", "Я иду домой.", "Торт вкусный.", "Он мой брат.", "Она очень умная.",
    "Это моя любимая песня.", "Мне нравится читать романы."
]

# Store them in a dictionary for easy access
data = {
    "en": english_sentences,
    "uz": uzbek_sentences,
    "ru": russian_sentences
}

data_lengths = {lang: len(sents) for lang, sents in data.items()}
data_lengths
```

```python
texts = english_sentences + uzbek_sentences + russian_sentences
labels = ["en"] * len(english_sentences) + ["uz"] * len(uzbek_sentences) + ["ru"] * len(russian_sentences)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X = vectorizer.fit_transform(texts)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
```

```python
# Train/Test bo'lish
X_train, X_test, y_train, y_test = train_test_split(X, labels, test_size=0.3, random_state=42)
```

```python
# SVM modeli
model = SVC(kernel='linear')
model.fit(X_train, y_train)
```

```python
# Baholash
y_pred = model.predict(X_test)
print("Aniqlik (accuracy):", accuracy_score(y_test, y_pred))
```

```python
# PCA bilan 2D ga tushiramiz
X_dense = X.toarray()
pca = PCA(n_components=2)
X_2d = pca.fit_transform(X_dense)
```

```python
colors = {"en": "blue", "uz": "green", "ru": "red"}
plt.figure(figsize=(8, 6))
for i, label in enumerate(labels):
    plt.scatter(X_2d[i, 0], X_2d[i, 1], color=colors[label], label=label if label not in plt.gca().get_legend_handles_labels()[1] else "")
plt.title("Language Classification (SVM + PCA)")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.legend()
plt.show()
```
